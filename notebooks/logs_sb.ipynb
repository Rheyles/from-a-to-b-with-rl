```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import json
```

##### Listing the available files

**Note** : Some of them will be labeled "continuous", they use the continuous model.

```python
files = glob.glob('../logs/a2c/*')
for no, file in enumerate(files):
    fname = file.replace('_', ' ')
    print(f"{no:2d} | {fname}")
```

```python
choice = -1 # Make your choice here
```

##### Plotting the data

```python
fig, axes = plt.subplots(nrows=3, sharex=True, figsize=[8,7])

csvs = glob.glob(files[choice] + '/*.csv' )
for csv in csvs:
    df = pd.read_csv(csv, skiprows=[0])
    df.head()

    axes[0].plot(df['r'], alpha=0.285)
    axes[0].plot([0,3500], [0,0], ':k')
    axes[0].plot(df['r'].rolling(50, center=True).mean())
    axes[1].plot(df['l'], color='orchid')
    axes[2].semilogy(df['t'], color='slateblue')
    
```
